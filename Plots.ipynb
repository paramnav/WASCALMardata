## Import 

```python
import matplotlib.pyplot as plt
import pandas as pd
```

## Input of Data

```python
path = '/home/jovyan/shared_data/data/aeronet/20210101_20241231_Mindelo_OSCM/20210101_20241231_Mindelo_OSCM.lev20'
data = pd.read_csv(path, sep = ',', skiprows = 6)

ODs = data.iloc[:, 4:26]

list(ODs.columns)
```

## Plots

```python
plt.plot(ODs['AOD_1640nm'].where(ODs['AOD_1640nm']>0))
plt.plot(ODs['AOD_500nm'].where(ODs['AOD_500nm']>0))
plt.legend(['1640nm', '500nm'])
plt.ylabel('Optical Density')
plt.xlabel('Measurements Events')
plt.title('Optical Density at 1640 nm')
```
