```python
import matplotlib.pyplot as plt
```

```python
df.insert(2, "Date Time ISO",pd.to_datetime(df["Date(dd:mm:yyyy)"] + " " +df["Time(hh:mm:ss)"], format='%d:%m:%Y %H:%M:%S') , True)
df = df.where(df != -999.0, np.nan) 
```

```python
df.iloc[:, 5:28]
```

```python
df.set_index("Date Time ISO").filter(like = "AOD").resample("6D").median().plot(legend = False)
```

```python
plt.figure(figsize = (15, 5))
plt.plot(df['Date Time ISO'], df['AOD_500nm'], 'x', linewidth=0.5)
plt.plot(df['Date Time ISO'], df['AOD_1640nm'], 'x',linewidth=0.5)
plt.legend(['500nm', '1640nm'])
plt.ylabel('Optical Density')
plt.xlabel('Time')
plt.title('Optical Density at 1640 nm and 500 nm')
```

```python
df = df.drop(['AOD_1640nm',
 'AOD_865nm',
 'AOD_779nm',
 'AOD_667nm',
 'AOD_620nm',
 'AOD_560nm',
 'AOD_555nm',
 'AOD_551nm',
 'AOD_532nm',
 'AOD_531nm',
 'AOD_510nm',
 'AOD_500nm',
 'AOD_490nm',
 'AOD_443nm',
 'AOD_412nm',
 'AOD_400nm',
 'AOD_380nm',
 'AOD_340nm'], axis = 'columns')
```

```python
wavelength = list(df.iloc[:, 5:9])
wavelength
```

```python
wavelength_int = []
for item in wavelength:
    wl = item.split('_')[-1]
    wl = wl[0:-2]
    wavelength_int.append(int(wl))

wavelength_int = np.array(wavelength_int)
wavelength_int
```

```python
plt.figure(figsize = (15, 5))
plt.plot(wavelength_int, df.iloc[1, 5:9], 'x-')
plt.plot(wavelength_int, df.iloc[5000, 5:9], 'x-')
plt.ylabel('Optical Density')
plt.xlabel('Wavelength')
plt.title('Optical Density over wavelength')
```

```python
AOD_675nm_df = df[['Date Time ISO', 'AOD_675nm']].set_index(['Date Time ISO']).resample('2H').mean()
AOD_675nm_df
```

```python
plt.figure(figsize = (15,5))
plt.plot(AOD_675nm_df['AOD_675nm'])
plt.ylabel('Optical Density')
plt.xlabel('Time')
plt.title(' Dust over Time')
```
